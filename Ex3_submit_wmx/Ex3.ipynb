```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
import matplotlib as plt

ts = pd.read_csv('data1forEx1to4/test1_icu_data.csv')
tr = pd.read_csv('data1forEx1to4/train1_icu_data.csv')
Y_tr = pd.read_csv('data1forEx1to4/train1_icu_label.csv')
Y_ts = pd.read_csv('data1forEx1to4/test1_icu_label.csv')
# ts = pd.read_csv('../Ex1_20210923/data1forEx1to4/test1_icu_data.csv')
# tr = pd.read_csv('../Ex1_20210923/data1forEx1to4/train1_icu_data.csv')
# Y_tr = pd.read_csv('../Ex1_20210923/data1forEx1to4/train1_icu_label.csv')
# Y_ts = pd.read_csv('../Ex1_20210923/data1forEx1to4/test1_icu_label.csv')


x_ts = np.array(ts)
x_tr = np.array(tr)
y_ts = np.array(Y_ts).ravel()
y_tr = np.array(Y_tr).ravel()
```

```python
X_train, X_test, y_train, y_test =train_test_split(x_tr,y_tr, train_size=0.8)

ker_list = ('rbf','linear','poly')
C_list =[0.001,1]#[0.001,0.1,1,10]
gamma_list = [0.001,1]#[0.001,0.1,1,10]
param_grid = {'kernel':ker_list,'gamma': gamma_list,'C':C_list}
clf = GridSearchCV(SVC(), param_grid, cv=5,n_jobs=4, return_train_score=True)
clf.fit(X_train, y_train)
print("best param: {0}\nbest score: {1}".format(clf.best_params_,clf.best_score_))
clf.best_estimator_
```

Since the poor computing power of my laptop, only 3\*2\*2-2=10 sets of parameters are tried.

```python
for params, acc_tr, acc_ts in zip(clf.cv_results_['params'],clf.cv_results_['mean_train_score'],clf.cv_results_['mean_test_score']):
    print('\n{}\n train accuracy:{:0.5f}, test accuracy:{:0.5f}'.format(params, acc_tr, acc_ts))
```

```python
best_point = list(clf.best_params_.values())
svm = SVC(C=best_point[0],gamma=best_point[1],kernel=best_point[2])
svm.fit(x_tr, y_tr)
pred_tr = svm.predict(x_tr)
pred_val = svm.predict(x_ts)

from sklearn.metrics import accuracy_score
train_acc = accuracy_score(y_tr, pred_tr)
vali_acc = accuracy_score(y_ts, pred_val)
print('train acc:{:0.5f}\nvalidation acc:{:0.5f}'.format(train_acc, vali_acc))
```
