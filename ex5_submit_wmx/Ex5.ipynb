### Part 11: Bayesian Network

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pgmpy.models import NaiveBayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from scipy.stats import kstest, shapiro

ts = pd.read_csv('data1forEx1to4/test1_icu_data.csv')
tr = pd.read_csv('data1forEx1to4/train1_icu_data.csv')
Y_tr = pd.read_csv('data1forEx1to4/train1_icu_label.csv')
Y_ts = pd.read_csv('data1forEx1to4/test1_icu_label.csv')
# ts = pd.read_csv('../Ex1_20210923/data1forEx1to4/test1_icu_data.csv')
# tr = pd.read_csv('../Ex1_20210923/data1forEx1to4/train1_icu_data.csv')
# Y_tr = pd.read_csv('../Ex1_20210923/data1forEx1to4/train1_icu_label.csv')
# Y_ts = pd.read_csv('../Ex1_20210923/data1forEx1to4/test1_icu_label.csv')


x_ts = np.array(ts)
x_tr = np.array(tr)
y_ts = np.array(Y_ts).ravel()
y_tr = np.array(Y_tr).ravel()
print('train:{}\ntest:{}'.format(x_tr.shape,x_ts.shape))
```

Check whether each feature follows Gaussian distribution with K-S test.

```python
def ks_check(x,cdf='norm'):
    pvalue = np.zeros((1,x.shape[1]))
    for i in range(x.shape[1]):
        _,pvalue[0,i]=kstest(x[:,i],cdf=cdf)
    return pvalue

def w_check(x):
    pvalue = np.zeros((1,x.shape[1]))
    for i in range(x.shape[1]):
        _,pvalue[0,i]=shapiro(x[:,i])
    return pvalue

p_tr = ks_check(x_tr,cdf='norm')
p_ts = ks_check(x_ts,'norm')

```

Obviously the features don\'t follow normal distriburion.

I tried [`sklearn.naive_bayes.GaussianNB`](https://scikit-learn.org/stable/modules/naive_bayes.html#gaussian-naive-bayes) ~~and [`pgmpy.models.NaiveBayes`](https://pgmpy.org/models/naive.html#module-pgmpy.models.NaiveBayes)~~ for classifier construction.

```python
clf_1 = GaussianNB().fit(x_tr, y_tr)
pred_tr_1 = clf_1.predict(x_tr)
pred_ts_1 = clf_1.predict(x_ts)
train_acc_1 = accuracy_score(y_tr, pred_tr_1)
test_acc_1 = accuracy_score(y_ts, pred_ts_1)
print('sklearn-Gaussian Naive Bayes\ntrain acc:{:0.5f}\nvalidation acc:{:0.5f}'.format(train_acc_1, test_acc_1))
print('train err:{:0.5f}\nvalidation err:{:0.5f}'.format(1-train_acc_1, 1-test_acc_1))
```
